## Множество Жюлиа 

Для фиксированного комплексного числа **c**, множество Жюлиа **J(c)** — это множество комплексных чисел $\mathbf{z_0}$, для которых последовательность
$$
z_{i+1} = z_i^2 + c 
$$
расходится.

## 1. Постройте двумерный график (по осям **Re(z), Im(z)**), при фиксированном **c** (в примере 0.1+0.5i)  с использованием 2 цветов (один цвет для сходящейся точки, другой — для расходящейся).

````python
### TODO:
### Постройте двумерный график (по осям **Re(z), Im(z)**), при фиксированном **c** (в примере 0.1+0.5i)
### с использованием 2 цветов (один цвет для сходящейся точки, другой — для расходящейся).


def iterate_until_nan(z_0: complex, c: complex, max_iterations: int = 20):
    """Iteration process to see if z_0 converges or diverges
    Args:
        z_0: initial value of z
        c: value of c
        max_iterations: number of maximum iterations
    Returns:
        int: zero if the iteration diverged,
             one if we reach max_iterations and we still
             haven't reached nan
    ### А почему на буржуйском языке?)
    """
    # РЕАЛИЗОВАТЬ ВОТ ТУТ
    ### Определяем, расходится ли последовательность z(n+1) = z(n^2) + c
    z = z_0
    for i in range(max_iterations):
        if abs(z) > 2:  # если модуль больше 2, то считаем что расходится
            return 0  # значит, расходится
        z = z * z + c
    return 1  # значит, не расходится


def julia_convergence(c, extent=[-2, 2, -1.5, 1.5], resulotion=0.01):
    W = int(abs(extent[1] - extent[0]) / resulotion) + 1
    H = int(abs(extent[2] - extent[3]) / resulotion) + 1
    img = np.empty((H, W))
    for i, x in enumerate(np.linspace(extent[0], extent[1], W)):
        for j, y in enumerate(np.linspace(extent[2], extent[3], H)):
            z_0 = complex(x, y)
            img[j, i] = iterate_until_nan(z_0=z_0, c=c)
    return img
````

````python
plt.figure(figsize=(25, 5))
plt.imshow(
    julia_convergence(0.1 + 0.5j),
    extent=[-2, 2, -1.5, 1.5],
    cmap="viridis",  # Pastel1  viridis Reds Spectral YlOrRd YlGnBu_r CMRmap cividis magma gnuplot gist_stern  plasma
    origin="lower",
)

plt.xlabel("Re(z)")
plt.ylabel("Im(z)")
plt.title("Множество Жюлия")
plt.colorbar(label="1 - сходится, 0 - расходится")
plt.show()
````

## 2. Придумайте логику из более чем двух цветов и нарисуйте картинку

````python
def julia(z_0: complex, c: complex = 1, n_iterations: int = 20):
    """Iteration process much like iterate_until_nan
       but you need think of some logic to give different
       values of z different colors
    Args:
        z_0: initial value of z
        c: value of c
        max_iterations: number of maximum iterations
    Returns:
        int
    """
    # РЕАЛИЗОВАТЬ ВОТ ТУТ
    z = z_0
    for i in range(n_iterations):
        if abs(z) > 2:
            return i  # если вышли за предел, возвращаем количество итераций
        z = z * z + c
    return n_iterations  # если не вышли, возвращаем max


def julia_set(c, extent=[-2, 2, -1.5, 1.5], resulotion=0.01):
    W = int(abs(extent[1] - extent[0]) / resulotion) + 1
    H = int(abs(extent[2] - extent[3]) / resulotion) + 1
    img = np.empty((H, W))
    for i, x in enumerate(np.linspace(extent[0], extent[1], W)):
        for j, y in enumerate(np.linspace(extent[2], extent[3], H)):
            z_0 = complex(x, y)
            img[j, i] = julia(z_0=z_0, c=c)
    return img
````

## 3. Нарисуйте множество Мандельброта (как в Википедии)
    В чем разница?

````python
# Напишите здесь код
"""
z_0 = 0, 
z_k+1 = z_k2 + c, 
k = 0, 1, 2,…
"""


def mandelbort(c: complex, max_iters: int = 20):
    """
    Функция рассчитывает, через сколько итераций точка
    уходит на расстояние больше 2 от начала координат.
    Args:
        - c: complex
        - max_iters: int (Максимальное число итераций)
    Returns:
        - i: int
    """
    z = 0
    for i in range(max_iters):
        if abs(z) > 2:
            return i
        z = z * z + c
    return max_iters


def julia_set2(c, extent=[-2, 2, -1.5, 1.5], resulotion=0.01, max_iters: int = 20):
    W = int(abs(extent[1] - extent[0]) / resulotion) + 1
    H = int(abs(extent[2] - extent[3]) / resulotion) + 1
    img = np.empty((H, W))
    for i, x in enumerate(np.linspace(extent[0], extent[1], W)):
        for j, y in enumerate(np.linspace(extent[2], extent[3], H)):
            c = complex(x, y)
            img[j, i] = mandelbort(c=c)
    return img
````

````python
plt.figure(figsize=(25, 5))
plt.imshow(
    julia_set2(c=(-2, 1, -1.5, 1.5), max_iters=50),
    ### Методом перебора искал. Возможно X*j должен быть равен 0, а не x->стремящийся к бесконечности
    ### Я к сожалению не разобрал аспект получения верной формы
    ### Возможно есть мат формула для описания формы объекта
    extent=[-2, 2, -1.5, 1.5],
    cmap="bone",  # Pastel1  viridis Reds Spectral YlOrRd YlGnBu_r CMRmap cividis magma gnuplot gist_stern  plasma
    origin="lower",
)

plt.xlabel("Re(z)")
plt.ylabel("Im(z)")
plt.title("Множество Мандельброта")
plt.colorbar(label="1 - сходится, 0 - расходится")
plt.show()
````

## 2. В файле script.py определены 8 функций. Построить график, наглядно демонстрирующий сколько времени проводится внутри каждой из этих 8 функций. Если вызывается функция, которая вложена в другую, то время исполнения засчитывается обеим функциям

````python
import script as picasso_draw_new_pictute
````

````python
import cProfile
import pstats
import io
import matplotlib.pyplot as plt

def profile_functions():
    """
    Запускаем cProfile для оценки времени выполнения функций.
    Return:
        - ps: Stats
    """
    profiler = cProfile.Profile()
    profiler.enable()

    ### Запускаем script.py
    picasso_draw_new_pictute.main()

    profiler.disable()

    s = io.StringIO()
    ps = pstats.Stats(profiler, stream=s).sort_stats("cumtime")
    ps.print_stats()

    print(s.getvalue())
    return ps


def plot_profile_stats(ps):
    """
    Создаем график времени выполнения заданных функций.
    """
    funcs_to_track = [
        "generate_domains",
        "initial_conditions",
        "add_perturb_1D",
        "add_noize",
        "non_linear_step",
        "computation_loop",
        "iterations",
        "plot_1D_heatmap",
    ]
    stats = ps.stats
    func_times = {}
    for func_desc, stat in stats.items():
        funcname = func_desc[2]
        if funcname in funcs_to_track:
            func_times[funcname] = stat[3]  # cumtime

    for f in funcs_to_track:
        if f not in func_times:
            func_times[f] = 0

    sorted_funcs = sorted(func_times.keys())
    times = [func_times[f] for f in sorted_funcs]

    plt.figure(figsize=(15, 7))
    plt.bar(sorted_funcs, times, color="skyblue")
    plt.ylabel("Время (сек)")
    plt.xlabel("Функции")
    plt.title("Время выполнения функций (профилировщик - Profile)")
    # plt.show()
    plt.savefig(f'task_2.2.png', dpi=300)
````

````python
ps = profile_functions()
````

````python
plot_profile_stats(ps)
````
